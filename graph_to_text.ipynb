```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('Screenshot 2023-10-25 140904.png')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Detect edges using the Canny edge detector
edges = cv2.Canny(blurred, 50, 150, apertureSize=3)

# Find contours in the edges
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Filter for rectangles by checking the number of vertices
rectangles = []
for contour in contours:
    perimeter = cv2.arcLength(contour, True)
    vertices = cv2.approxPolyDP(contour, 0.04 * perimeter, True)
    if len(vertices) == 4:
        rectangles.append(vertices)

# Draw and save each rectangle as a separate image
for i, rect in enumerate(rectangles):
    # Create a mask for the current rectangle
    mask = np.zeros_like(image)
    cv2.drawContours(mask, [rect], -1, (255, 255, 255), thickness=cv2.FILLED)

    # Bitwise-AND to extract the rectangle
    result = cv2.bitwise_and(image, mask)

    # Save the rectangle as a separate image
    cv2.imwrite(f'rectangle_{i}.jpg', result)

# Display or save the image with rectangles highlighted
cv2.drawContours(image, rectangles, -1, (0, 255, 0), 2)
cv2.imshow('Image with Rectangles', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

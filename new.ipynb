```python
img = Image("full_page.jpg")
tesseract = TesseractOCR()

# Extract tables with Tesseract and PaddleOCR
tables = img.extract_tables(ocr=tesseract, borderless_tables=True)

df = tables[0].df
print(df)
```

```python
#remove first row
df = df.iloc[1:]
```

```python
# Extract the first column as an array (list) except the header
row_list = df.iloc[0:, 0].values
# print(row_list)
#remove the first element
print(row_list)
```

```python
import pandas as pd

# Convert the table to a DataFrame
table_df = pd.DataFrame(tables[0].df)

# Now you can work with 'table_df' as a DataFrame
# print(table_df)

#for all rows, check if the values expcept first column is empyt, if it is print that this row is empty
empty_rows = []
for index, row in table_df.iterrows():
    # print(row)
    if row[1:].isnull().all():
        print("Row {} is empty".format(index))
        empty_rows.append(index)
        
print(empty_rows)
```

```python
for(index, row) in table_df.iterrows():
    #find the row with index in empty_rows
    if index in empty_rows:
        row_name = row_list[index-1] 
        next = index+1
        #appent the row name to the next row
        table_df.iloc[next, 0] = row_name  + table_df.iloc[next, 0]
    
print(table_df)
```

```python
#remove the rows whose index is in empty_rows
table_df = table_df.drop(empty_rows)
print(table_df)
```

```python
#remove the first row
table_df = table_df.iloc[1:]
print(table_df)

```

```python
    
#if value of row in first column is none, then remove row



print(table_df)
```

```python
#remove the column which have all values as none
table_df = table_df.dropna(axis=1, how='all')
print(table_df)

```

```python
#remove the rows whose element in forst column is none
table_df = table_df.dropna(subset=[table_df.columns[0]])
print(table_df)
```

```python
#reset the index
table_df = table_df.reset_index(drop=True)
```

```python
print(table_df)
```

```python
#remove \n from the first column elements
table_df[table_df.columns[0]] = table_df[table_df.columns[0]].str.replace('\n', ' ')
```
